## Creating Logistic Regression Class

```python
class logistic_regression():

    def sigmoid(self,x):
        return 1/(1+np.exp(-x))

    def __init__(self,learning_rate=0.01,no_of_iters=1000):
       ## initialize the instance variables
       self.learning_rate = learning_rate
       self.no_of_iters = no_of_iters
       self.th = None
       self.th0 = None
    
    def fit(self,X,Y):
       ## contains the training part 
        (n_samples,n_features) = X.shape
        self.th = np.zeros((n_features,1))
        self.th0 = 0
        
        for i in range(self.no_of_iters):

            linear_model = X @ self.th + self.th0
            y_pred  = self.sigmoid(linear_model).reshape(-1,1)

            # dth = ((1/n_samples) * (X.T @ (y_pred - Y)).T).reshape(-1

            dth = 1/n_samples * (X.T @ (y_pred - Y))
            dth0 = (1/n_samples) * np.sum(y_pred - Y)

            self.th = self.th - self.learning_rate * dth
            self.th0 = self.th0 - self.learning_rate * dth0  
            # print(i)    

    #  def update_weights(self):
    #    ## weight update step logistic regression

    def predict(self,X):
        ## predict the value using weights, biases and the input feature vector
        
        linear_model = X @ self.th + self.th0
        y_pred = self.sigmoid(linear_model).reshape(-1,1)
        
        y_pred_cls = np.array([1 if i > 0.5 else -1 for i in y_pred]).reshape(-1,1)
        return y_pred_cls
```

## Running the Classifier

Dividing the Data into training and testing

```python
data_train = data[:400]
data_test = data[400:]
labels_train = labels[:400]
labels_test = labels[400:]
```

```python
def accuracy(y_true,y_pred):
    accuracy = np.sum(y_true == y_pred) / len(y_true)
    return accuracy

def cross_validation(logistic_regression,data,labels,k=5):
    data_folds = np.array_split(data,k)
    labels_folds = np.array_split(labels,k)
    accuracies = []
    for i in range(k):
        data_train = np.concatenate([data_folds[j] for j in range(k) if j != i])
        labels_train = np.concatenate([labels_folds[j] for j in range(k) if j != i])
        data_test = data_folds[i]
        labels_test = labels_folds[i]
        logistic_regression.fit(data_train,labels_train)
        predictions = logistic_regression.predict(data_test)
        accuracies.append(accuracy(labels_test,predictions))
    return np.mean(accuracies)
```

```python
classifier  = logistic_regression()
classifier.fit(data_train,labels_train)
predictions = classifier.predict(data_test)
print(accuracy(labels_test,predictions))
```

```python
cross_validation(classifier,data,labels,10)
```
